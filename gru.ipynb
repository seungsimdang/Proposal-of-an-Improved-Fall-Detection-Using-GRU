```python
# 각각 R값, HSSC값 텍스트 파일은 열어서 읽어온다.
with open("/content/drive/MyDrive/Colab Notebooks/urfd values/Rvalue(30).txt", "r") as f:
  rvalue = f.read();

with open("/content/drive/MyDrive/Colab Notebooks/urfd values/HSSCvalue(30).txt", "r") as f:
  hsscvalue = f.read();

# 이때, 값들은 모두 문자열이고 ','로 구분이 되어 있기 때문에 split을 진행.
rList = rvalue.split(',')
hsscList = hsscvalue.split(',')
newRList = []
newHsscList = []

rList = rList[1:]
rList = rList[:len(rList)-2]

for rl in rList:
  if rl == '[null':
    i = rList.index('[null')
    del rList[i]
    del rList[i-1]
  real = round(float(rl),3)
  newRList.append(real)

hsscList = hsscList[1:]
hsscList = hsscList[:len(hsscList)-2]

for hl in hsscList:
  if hl == '[null':
    j = hsscList.index('[null')
    del hsscList[j]
    del hsscList[j-1]
  real = round(float(hl),3)
  newHsscList.append(real)

print(len(newRList))
print(len(newHsscList))
print(newRList)
print(newHsscList)
```
